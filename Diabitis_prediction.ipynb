```python
import pandas as pd
```

```python
dt = pd.read_csv("diabities.csv")
```

```python
dt.head()
```

```python
dt.info()
```

```python
x=dt.iloc[: , : -1]
y=dt.iloc[: , -1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators = 6, criterion = 'entropy')
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc_log = accuracy_score(y_test,y_pred)*100
print("Accuracy score is :",acc_log)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs',max_iter=1000)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
acc_log = accuracy_score(y_test,y_pred)*100
print("Accuracy score for linear regression is :",acc_log)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
acc_log = accuracy_score(y_test,y_pred)*100
print("Accuracy score for KNeighborsClassifier is :",acc_log)
```
